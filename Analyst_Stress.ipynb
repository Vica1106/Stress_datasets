```python
gene_counts = merged_df['Gene_Name'].value_counts().reset_index()
gene_counts.columns = ['Gene_Name', 'Count']
gene_pvalues = merged_df.groupby('Gene_Name')['padj'].mean().reset_index()
gene_counts = pd.merge(gene_counts, gene_pvalues, on='Gene_Name')

gene_urls = merged_df.groupby('Gene_Name')['URL'].first().reset_index() 
gene_counts = pd.merge(gene_counts, gene_urls, on='Gene_Name', how='left')
# Filter genes with count > 1
gene_counts_filtered = gene_counts[gene_counts['Count'] > 1]
gene_counts_filtered.to_csv('./Stress_datasets/TableCount_padj_UP_KW_DOMAIN.tsv', sep='\t', index=False)
gene_counts_filtered
```

```python
# Normalize and set up color mapping
norm = plt.Normalize(gene_counts_filtered['padj'].min(), gene_counts_filtered['padj'].max())
colors = cm.viridis(norm(gene_counts_filtered['padj']))

# Plot the bar chart with color mapping
fig, ax = plt.subplots(figsize=(10, 6))  # Explicitly create figure and axes
bars = ax.bar(gene_counts_filtered['Gene_Name'], gene_counts_filtered['Count'], color=colors)

# Set labels and title
ax.set_xlabel('Gene Name')
ax.set_ylabel('Count')
ax.set_title('Gene Counts with Padj-value-Based Coloring (Count > 1)')
ax.set_xticks(range(len(gene_counts_filtered['Gene_Name'])))
ax.set_xticklabels(gene_counts_filtered['Gene_Name'], rotation=45, ha='right')

# Set y-axis to only have integer values
ax.yaxis.set_major_locator(MaxNLocator(integer=True))

max_count = gene_counts_filtered['Count'].max()
ax.set_ylim(0, max_count + 0.5)

# Add a colorbar for reference
sm = cm.ScalarMappable(cmap='viridis', norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, orientation='horizontal', pad=0.2)
cbar.set_label('Adjusted P-value (Padj)')  # Changed label to be more descriptive

# Annotate p-values on bars
for i, pval in enumerate(gene_counts_filtered['padj']):
    ax.text(i, gene_counts_filtered['Count'].iloc[i] + 0.1, f'{pval:.2e}', ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```
